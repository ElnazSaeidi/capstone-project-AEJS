## EDA weather data - temperature

```python
#defining a new dataframe that contains the average temperature of all weather stations

#dropping columns
tavg_df_ire=weather_df_ire.drop(['weather_tmin','weather_tmax','weather_prcp','weather_snow','weather_wdir','weather_wspd','weather_wpgt','weather_pres','weather_tsun'], axis='columns')

tavg_df_can=weather_df_can.drop(['weather_tmin','weather_tmax','weather_prcp','weather_snow','weather_wdir','weather_wspd','weather_wpgt','weather_pres','weather_tsun'], axis='columns')

tavg_df_fran=weather_df_fran.drop(['weather_tmin','weather_tmax','weather_prcp','weather_snow','weather_wdir','weather_wspd','weather_wpgt','weather_pres','weather_tsun'], axis='columns')

#defining a large dataframe

tavg_df=pd.concat([tavg_df_ire,tavg_df_can,tavg_df_fran])

#creating an empty data frame including the weather_date to add the data of the other stations later

tavg_df_dublin=tavg_df_ire[tavg_df_ire['station']=='03969']
tavg_df_all=tavg_df_dublin.drop(['station'], axis='columns')
tavg_df_all

#filtering the complete data set for the weather stations
#adding the temperature data to the new dataset to have the temperature for each station in a column

stations=['03969','03962','03955','71863','71508','71877','07510','07059','07643']

for station in stations:
       tavg_df_temp=tavg_df[tavg_df['station']==station]
       tavg_df_all=pd.merge(tavg_df_all,tavg_df_temp, on='weather_date', suffixes=('', f'_{station}'))
       tavg_df_all=tavg_df_all.drop('station', axis='columns')

tavg_df_all=tavg_df_all.drop('weather_tavg', axis='columns')
tavg_df_all=tavg_df_all.reset_index()
tavg_df_all


```

```python
# defining a dataframe that shows the monthly temperature to have a smooth curve
# using groupby

tavg_df_all['weather_date'] = pd.to_datetime(tavg_df_all['weather_date'])

tavg_df_month=tavg_df_all.groupby(tavg_df_all['weather_date'].dt.to_period("M")).mean().round()

#through this operation, two indexes are built (weather_date)
#rename one of it 

tavg_df_month=tavg_df_month.rename(columns={'weather_date':'weather_date_day'})

#reset the index
tavg_df_month=tavg_df_month.reset_index()

#drop one of the weather date columns

tavg_df_month=tavg_df_month.drop('weather_date_day', axis='columns')

#renaming the columns by using a dictionary

city_station={
    '03969':'dublin',
    '03962':'shannon',
    '03955':'cork',
    '71863':'regina',
    '71508':'toronto',
    '71877':'calgary',
    '07510':'bordeaux',
    '07059':'amiens',
    '07643':'montpellier'
}

# dropping the 'weather_tavg_' from the column name

tavg_df_month.columns=tavg_df_month.columns.str.replace('weather_tavg_','')

tavg_df_month=tavg_df_month.rename(columns=city_station)
tavg_df_month.sample(5)




```

# Statistical calculations on the weather data

```python
# Did the temperature in Canada increase significantly in 2019 compared to 1980?
# Regarding the whole year

df_weather_canada.head(5)

#calculating the mean temperature  

tavg_q1=df_weather_canada[df_weather_canada['weather_date']<='1990-01-01']['weather_tavg']

tavg_q4=df_weather_canada[df_weather_canada['weather_date']>='2009-12-31']['weather_tavg']

tavg_q4.sample(4)

tavg_q1.dropna()
tavg_q4.dropna()


tavg_q1.to_csv(r'/Users/anjakurzhals/neuefische/capstone_local/canada_tavg_q1.csv', index=False)
tavg_q4.to_csv(r'/Users/anjakurzhals/neuefische/capstone_local/canada_tavg_q4.csv', index=False)
```

```python

from scipy.stats import ttest_ind

# Assuming group1 and group2 are your two groups
t_stat, p_value = ttest_ind(tavg_q1, tavg_q4)

print(p_value)

if p_value < 0.05:
    print("The groups differ significantly.")
else:
    print("There is no significant difference between the groups.")
```
